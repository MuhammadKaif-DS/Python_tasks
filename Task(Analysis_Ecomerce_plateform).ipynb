```python
import pandas as pd
```

```python
Orders = pd.read_csv('orders.csv')
Orders
```

```python
Products = pd.read_csv('products.csv') 
Products
```

```python
Customers = pd.read_csv('customers.csv')
Customers
```

```python
Orders_M = Orders.merge(Products[['product_id','product_name','category','price']], on = 'product_id', how = 'left')
```

```python
Orders_MF = Orders_M.merge(Customers[['customer_id','gender','age']], on = 'customer_id', how = 'left')
```

```python
Orders_MF.head()
```

```python
Orders_MF['Years'] = Orders_MF['order_date'].str.split('/').str[2].astype(int)
```

```python
Orders_MF['Months'] = Orders_MF['order_date'].str.split('/').str[0].astype(int)
```

```python
Orders_MF.info()
```

```python
grouped = Orders_MF.groupby('Years')
grouped.get_group(2022)
```

```python
def Age_limits(Age):
    if 8< Age <18:
        return('Tenager_limit')
    elif 18< Age <40:
        return('Adult_limit')
    elif Age >40:
        return('Old_limit')
    
```

```python
Orders_MF['Age_limit'] = Orders_MF.age.apply(Age_limits)
```

```python
Orders_MF
```

```python
Mo = Orders_MF.Months.unique()
Mo
```

```python
Ye = Orders_MF.Years.unique()
Ye
```

```python
def y_lest_sold(product):
    if  product == 3:
        return(product)
```

```python
def m_lest_sold(product):
    if  product == 1:
        return(product)
```

```python
def discount_customer(count):
    if count >5:
        return(count)
```

```python
def m_discout_customers(count):
    if count >5:
        return(count)
    else:
        return("None")
```

```python
for ye in Ye:
    print('*'*8,ye,'*'*8)
    g = Orders_MF.groupby('Years')
    ye_g = g.get_group(ye)
    Repeted_CI = ye_g.customer_id.value_counts()
    Top_selling_product = ye_g.product_name.value_counts()
    lest_s = ye_g.product_name.value_counts().apply(y_lest_sold).dropna()
    Discount_CI = ye_g.customer_id.value_counts().apply(discount_customer).dropna()
    Year_Revenue = (ye_g.quantity * ye_g.price).sum()
    print(f"-----Repeated_Coustomers_ID-----\n{Repeted_CI}\n-----Top_selling_product_names-----\n{Top_selling_product}\n-----Least_sold_product_names-----\n{lest_s}\n-----Discount_ID-----\n{Discount_CI}\n----------\n{ye}_Total_Revenue {Year_Revenue}")
```

```python
for ye in Ye:
    Creat_groups_year = Orders_MF.groupby('Years')
    Groups_year = Creat_groups_year.get_group(ye)
    print('*'*5,ye,'*'*5)
    for mo in Mo:
        Create_groups_months = Groups_year.groupby('Months')
        Groups_months = Create_groups_months.get_group(mo)
        print('*'*5,mo,'*'*5)
        Repeated_CI = Groups_months.customer_id.value_counts()
        Top_selling_products = Groups_months.product_name.value_counts()
        Least_sold_product = Groups_months.product_name.value_counts().apply(m_lest_sold).dropna()
        Discount_CI = Groups_months.customer_id.value_counts().apply(m_discout_customers).dropna()
        Months_Revenue = (Groups_months.quantity*Groups_months.price).sum()
        print(f"-----Repeated_ID-----\n{Repeated_CI}\n-----Top_selling_products-----\n{Top_selling_products}\n-----Least_sold_products-----\n{Least_sold_product}\n-----Discount_CI-----\n{Discount_CI}\n--------\n{mo}_Total_Revenue:{Months_Revenue}\n____________\n")
```
